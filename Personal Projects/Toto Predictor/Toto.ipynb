```python
# LOAD FILES

import pandas as pd
from datetime import datetime

toto = pd.read_csv("toto.csv")
all_time_toto = pd.read_csv("all_time_toto.csv")

month_list = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
day_list = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
system_acceptable_inputs = ('7','8','9','10','11','12','')
```

```python
# OOP FOR EACH WINNING DETAILS

class winning_details:
    def __init__(self,date,day,win1,win2,win3,win4,win5,win6,add,prize):
        self.date = date
        for month in month_list:
            if month in self.date:
                self.month = month
        self.year = self.date[-2:]
        self.day = day
        self.win1 = win1
        self.win2 = win2
        self.win3 = win3
        self.win4 = win4
        self.win5 = win5
        self.win6 = win6
        self.add = add
        self.prize = prize

# CREATE LIST OF WINNING ENTRY DETAILS AS OBJECTS FOR GIVEN YEAR CSV FILE
list_of_winning_details = []
for i in range(len(toto)):
    entry = toto.iloc[i,:]
    winning = winning_details(entry[0],entry[1],entry[2],entry[3],entry[4],entry[5],entry[6],entry[7],entry[8],entry[9])
    list_of_winning_details.append(winning)
```

```python
# COMMON

# SORTS DICTIONARY OF NUMBER : COUNT. DEFAULT SET TO HIGHEST COUNT TO LOWEST COUNT  &   RETURN 6 NUMBERS
def sort_numbers(num_dict,reverse = True,system=6):
    sorted_list = sorted(num_dict.items(), key=lambda x:x[1], reverse = reverse)
    return list(dict(sorted_list[:system]).keys())

# TO CREATE DICTIONARY OF NUMBERS
def update_dictionary(num,num_dict):
    if num not in num_dict.keys():
        num_dict[num] = 1
    else:
        num_dict[num] += 1
    return num_dict

# CONSOLIDATE ALL MAIN NUMBERS ONLY
def mass_update_main_numbers(winning_details,num_dict):
    update_dictionary(winning_details.win1,num_dict)
    update_dictionary(winning_details.win2,num_dict)
    update_dictionary(winning_details.win3,num_dict)
    update_dictionary(winning_details.win4,num_dict)
    update_dictionary(winning_details.win5,num_dict)
    update_dictionary(winning_details.win6,num_dict)
    return num_dict
```

```python
# MANUAL TOTO CONSOLIDATION FUNCTIONS

# DICTIONARY OF NUMBERS : MAIN BALL COUNT
def top_main_ball_numbers():
    num_dict = {}
    for winning_details in list_of_winning_details:
        mass_update_main_numbers(winning_details,num_dict)
    return num_dict

# DICTIONARY OF NUMBERS : ADDITIONAL BALL COUNT
def top_additional_numbers():
    num_dict = {}
    for winning_details in list_of_winning_details:
        update_dictionary(winning_details.add,num_dict)
    return num_dict

# DICTIONARY OF NUMBERS : TOTAL COUNT
def numbers_total_win_count():
    num_dict = {}
    main_ball_count_dict = top_main_ball_numbers()
    additional_count_dict = top_additional_numbers()
    for i in range(49):
        i += 1
        main_ball_count = 0
        additional_count = 0
        if i in main_ball_count_dict:
            main_ball_count = main_ball_count_dict[i]
        if i in additional_count_dict:
            additional_count = additional_count_dict[i]
        num_dict[i] = main_ball_count + additional_count
    return num_dict

# TOP NUMBERS, DEFAULT SYSTEM 6 / RETURNS 6-DIGITS
# PROVIDE INPUT PARAMETER IF WANT TO GET BETWEEN 7-12 DIGITS
def top_numbers(system=6):
    num_dict = numbers_total_win_count()
    return sort_numbers(num_dict,system=system)

def lowest_numbers(system=6):
    num_dict = numbers_total_win_count()
    return sort_numbers(num_dict,False,system=system)

# TOP NUMBERS BY DAY
def top_numbers_by_day(day,system=6):
    day = day.capitalize()[0:3]
    if day.capitalize()[0:3] not in day_list:
        exit
    num_dict = {}
    for winning_details in list_of_winning_details:
        if winning_details.day == day:
            mass_update_main_numbers(winning_details,num_dict)
            update_dictionary(winning_details.add,num_dict)
    return sort_numbers(num_dict,system=system)

# TOP NUMBERS BY MONTH
def top_numbers_by_month(month,system=6):
    if month.capitalize()[0:3] not in month_list:
        exit
    month = month.capitalize()[0:3]
    num_dict = {}
    for winning_details in list_of_winning_details:
        if winning_details.month == month:
            mass_update_main_numbers(winning_details,num_dict)
            update_dictionary(winning_details.add,num_dict)
    return sort_numbers(num_dict,system=system)

# TOP NUMBERS BY YEAR
def top_numbers_by_year(year,system=6):
    if year == "" or not(year.isdigit()) or len(year) < 2 or len(year) > 4 or int(year) > datetime.now().year:
        exit
    year = year[-2:]
    num_dict = {}
    for winning_details in list_of_winning_details:
        if winning_details.year == year:
            mass_update_main_numbers(winning_details,num_dict)
            update_dictionary(winning_details.add,num_dict)
    return sort_numbers(num_dict,system=system)

# TOP NUMBERS BY GRP 1 PRIZE
def numbers_by_prize(prize_amount,system=6):
    if prize_amount != int:
        exit
    num_dict = {}
    for winning_details in list_of_winning_details:
        if winning_details.prize >= prize_amount:
            mass_update_main_numbers(winning_details,num_dict)
            update_dictionary(winning_details.add,num_dict)
    return sort_numbers(num_dict,system=system)

# TOP NUMBER IN LAST X WEEKS

# ASSOCIATIVE NUMBERS
```

```python
# ALL TIME TOTO FUNCTIONS

class toto_numbers:
    def __init__(self,number,main_ball,additonal):
        self.number = number
        self.main_ball = main_ball
        self.additional = additonal

# CREATE LIST OF NUMBERS AS OBJECTS FROM CSV FILE
all_time_stats = []
for i in range(len(all_time_toto)):
    entry = all_time_toto.iloc[i,:]
    number = toto_numbers(entry[0],entry[1],entry[2])       # WHERE entry 1/2/3 ARE COLUMN NAMES : NUMBER / MAIN BALL COUNT / ADDITIONAL NUMBER COUNT
    all_time_stats.append(number)

# CREATE DICTIONARY OF NUMBERS : MAIN BALL COUNT
def all_time_top_main_ball_numbers():
    num_dict_main_ball_only = {}
    for number in all_time_stats:
        num_dict_main_ball_only[number.number] = number.main_ball
    return sort_numbers(num_dict_main_ball_only)

# CREATE DICTIONARY OF NUMBERS : ADDITIONAL BALL COUNT
def all_time_top_additional_numbers():
    num_dict_additional_only = {}
    for number in all_time_stats:
        num_dict_additional_only[number.number] = number.additional
    return sort_numbers(num_dict_additional_only)

# CREATE DICTIONARY OF NUMBERS : TOTAL COUNT
def overall_num_dict():
    num_dict = {}
    for number in all_time_stats:
        num_dict[number.number] = (number.main_ball + number.additional)
    return num_dict

# DEFAULT SYSTEM 6 / RETURNS 6 DIGITS
# PROVIDE INPUT PARAMETER TO RETURN 7-12 DIGITS
def all_time_top_numbers(system=6):
    num_dict = overall_num_dict()
    return sort_numbers(num_dict,system=system)

def all_time_lowest_numbers():
    num_dict = overall_num_dict()
    return sort_numbers(num_dict,False)
```

```python
# COLLECT INPUT DATA
day = input("Which day are you betting? (Mon or Thu)")
month = input("Which month are you betting? (E.g Jan)")
year = input("Which year's stats would you like to see?")
currentyear = str(datetime.now().year)
system = input("What system are you playing? (Input a number between 7-12, else leave blank)")

# CLEAN INPUT DATA
while system not in system_acceptable_inputs:
    system = input("What system are you playing? (Input a number between 7-12, else leave blank)")
if system != '':
    system = int(system)
prize_amount = input("What is the minimum group 1 prize amount that you are looking at?")

if prize_amount.isdigit():
    prize_amount = int(prize_amount)
else:
    prize_amount = ""
```

```python
# OVERALL ANALYSIS
print("******* All Time TOTO Stats *******")
print(f"All time top 6 numbers are {all_time_top_numbers()}")
print(f"All time lowest 6 numbers are {all_time_lowest_numbers()}")
print(f"Most common main ball numbers are {all_time_top_main_ball_numbers()}")
print(f"Most common additional numbers are {all_time_top_additional_numbers()}")
if system != "":
    print(f"Best numbers for System{system} are {all_time_top_numbers(system)}")

print(f"\n\n******* Manual TOTO Stats *******")
print(f"Top 6 numbers are {top_numbers()}")
print(f"Lowest 6 numbers are {lowest_numbers()}")
print(f"Most common main ball numbers are {sort_numbers(top_main_ball_numbers(),True)}")
print(f"Most common additional numbers are {sort_numbers(top_additional_numbers(),True)}")

if top_numbers_by_day(day):
    print(f"\n>>>>>>>  Stats for {day}")
    print(f"Best numbers for {day} are {top_numbers_by_day(day)}")

if top_numbers_by_month(month):
    print(f"\n>>>>>>>  Stats for {month}")
    print(f"Best numbers for {month} are {top_numbers_by_month(month)}")

if top_numbers_by_year(currentyear):
    print(f"\n>>>>>>>  Stats for Current Year {currentyear}")
    print(f"Best numbers for {currentyear} are {top_numbers_by_year(currentyear)}")

if prize_amount != "":
    print(f"\n>>>>>>>  Stats for Group 1 Prize More Than ${prize_amount}m")
    print(f"Best numbers are {numbers_by_prize(prize_amount,system=6)}")

if system != '':
    print(f"\n>>>>>>>  Stats for System{system}")
    print(f"Best numbers for System{system} are {top_numbers(system)}")
    # EACH IF STATEMENT VERIFIES IF THERE IS A SUCCESSFUL OUTPUT BEFORE DISPLAYING
    if top_numbers_by_month(month,system):
        print(f"Best numbers for System{system} in {month} are {top_numbers_by_month(month,system)}")
    if top_numbers_by_day(day,system):
        print(f"Best numbers for System{system} on {day} are {top_numbers_by_day(day,system)}")
    if top_numbers_by_year(currentyear,system):
        print(f"Best numbers for System{system} for current year {currentyear} are {(top_numbers_by_year(currentyear,system))}")
    if prize_amount != "":
        print(f"Best {system} numbers for Group 1 Prize Amount More Than ${prize_amount}m are {numbers_by_prize(prize_amount,system)}")
    
```
